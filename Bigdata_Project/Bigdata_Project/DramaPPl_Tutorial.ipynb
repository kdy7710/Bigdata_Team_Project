# API 호출 for 문 

```python

import pandas as pd
import date_cal as dc 
from Naver_Api.API_CLASS import  NaverApi # NaverApi 클래스 임포트
from Text_Mining import SetKorean # 그래프 한글 설정 
import time 
df = pd.read_excel('C:/Users/tjoeun/Downloads/7. 데이터 활용 매뉴얼 - 방송정보메타데이타샘플.xls')
# 사용할 컬럼만 추출 
df2 = df.loc[:,['프로그램','방송일','상품명','품번']]
# print(len(df2))
# print(df2)

# print(df2.loc[0,'상품명'])
final_df = pd.DataFrame()
for i in range(len(df2)):
    time.sleep(1)
    keyword = df2.loc[i,'상품명']
    broadcast_date = df2.loc[i,'방송일']
    date = dc.convert_strtime(broadcast_date)
    enddate = dc.timeminus(date, 30)
    startdate = dc.timeminus(date, -30)
    print(keyword,broadcast_date,date,enddate,startdate)
    api = NaverApi(keyword,startdate=startdate,enddate=enddate).to_dataframe()

    rate = list(api['비율_x'])
    
    final_df[keyword] = rate

    
   

print(final_df)
```
