### 테이블 수정 (날짜 집어넣기)

```python
import pandas as pd
```

```python
df = pd.read_excel('ppl_all_product.xlsx')
df.head()
```

```python
product = []
a = -30
for i in range(0,len(df['글리셜 미셀라 클렌징 워터'])):
    b = a+i
    if b < 0:
        product.append('D'+str(a+i))
    elif b == 0:     
        product.append('D-day')
    else:
        product.append('D+'+str(a+i))
product
```

```python
for i in product:
    sql = "ALTER TABLE `brand` ADD COLUMN `"+i+"` FLOAT NOT NULL DEFAULT 0 "
    cur.execute(sql)
```
